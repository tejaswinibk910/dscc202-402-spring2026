```python
# =============================================================================
# Data Quality Views
# =============================================================================
from pyspark.pipelines import materialized_view, table
from pyspark.sql import functions as F

@materialized_view(comment="Daily order summary for monitoring")
def daily_order_summary():
    """Daily aggregation of orders by region for monitoring"""
    return (spark.read.table("fact_orders")
            .groupBy(F.to_date("order_date").alias("order_date"), "region")
            .agg(
                F.count("*").alias("order_count"),
                F.sum("amount").alias("total_revenue"),
                F.countDistinct("customer_id").alias("unique_customers")
            ))
```
